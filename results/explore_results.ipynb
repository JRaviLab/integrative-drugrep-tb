```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
p = pd.read_csv('rna_ranks_p.csv', index_col=0)
s = pd.read_csv('rna_ranks_s.csv', index_col=0)
r = pd.read_csv('rna_ranks_r.csv', index_col=0)
l = pd.read_csv('rna_ranks_l.csv', index_col=0)

cols = np.random.choice(p.columns, 10, replace=False)

for col in cols:
    for i, data in enumerate([p,s,r,l]):
        print(col, i, data[col].sort_values(ascending=False).index.values[:5])
    print('\n')
```

```python
p = pd.read_csv('rna_ranks_p.csv', index_col=0)
s = pd.read_csv('rna_ranks_s.csv', index_col=0)
r = pd.read_csv('rna_ranks_r.csv', index_col=0)
l = pd.read_csv('rna_ranks_l.csv', index_col=0)

# cols = np.random.choice(p.columns, 10, replace=False)
cols = p.columns

scores = [0,0,0,0]
for col in cols:
    top_cells = []
    for i, data in enumerate([p,s,r,l]):
        top_cells.append(data[col].sort_values(ascending=False).index.values[0])
    for i in range(4):
        # if the values at top_cells[i] exists at least twice in top_cells, add one to score[i]
        if top_cells.count(top_cells[i]) > 1:
            scores[i] += 1 
    # print(top_cells)
print(len(cols))
scores

```
