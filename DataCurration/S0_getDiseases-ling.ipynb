#### Helper Function

```python
def split_and_clean_series_ids(series_list):
    """
    Split comma-separated series IDs and clean whitespace
    
    Parameters:
    series_list: List of strings that may contain comma-separated values
    
    Returns:
    numpy.ndarray: Array of cleaned, individual series IDs
    """
    split_list = []
    for item in series_list:
        split_list.extend([x.strip() for x in str(item).split(',')])
    
    return np.array(split_list)
```
