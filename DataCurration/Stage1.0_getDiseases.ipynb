#### Helper Function

```python
def split_and_clean_series_ids(series_list):
    """
    Split comma-separated series IDs and clean whitespace
    
    Parameters:
    series_list: List of strings that may contain comma-separated values
    
    Returns:
    numpy.ndarray: Array of cleaned, individual series IDs
    """
    split_list = []
    for item in series_list:
        split_list.extend([x.strip() for x in str(item).split(',')])
    
    return np.array(split_list)
```

**Mircroarrays data**

We queried the refine.bio database using the keywords “pulmonary tuberculosis,” “active tuberculosis,” and “MTB” to retrieve available gene expression datasets and their corresponding metadata. These datasets were originally deposited in the Gene Expression Omnibus (GEO) and were uniformly processed by [refine.bio](https://www.refine.bio/) up to June 2025. Data were downloaded according to the instructions provided on the refine.bio platform.

The datasets included in this study, along with detailed curation information, are listed in
`../data/labeling/TB_curationdoc.tsv.`

**RNAseq data**

Similarly, for RNA-seq data, we obtained uniformly processed gene expression data and associated metadata using the [archs4py](https://github.com/MaayanLab/archs4py) Python library with the same keywords: “pulmonary tuberculosis,” “active tuberculosis,” and “MTB.” Datasets available up to June 2025 were retrieved. Further details of the data retrieval process are provided in the code section below.

The datasets included in this study, together with detailed curation information, are listed in
`../data/labeling/TB_curationdoc.tsv.`

```python
# get sample meta data based on search term
PTB_meta = a4.meta.meta(file, "pulmonary tuberculosis", meta_fields=["geo_accession","series_id","characteristics_ch1", "extract_protocol_ch1", "source_name_ch1", "title"])

# get sample meta data based on search term
activeTB_meta = a4.meta.meta(file, "active tuberculosis", meta_fields=["geo_accession","series_id","characteristics_ch1", "extract_protocol_ch1", "source_name_ch1", "title"])

# get sample meta data based on search term
MTB_meta = a4.meta.meta(file, "MTB", meta_fields=["geo_accession","series_id","characteristics_ch1", "extract_protocol_ch1", "source_name_ch1", "title"])
```

```python
import pandas as pd
TB_meta = pd.concat([PTB_meta, activeTB_meta, MTB_meta], ignore_index=True)
TB_meta
```

```python
MTB_meta2 = a4.meta.meta(file, "mycobacterium tuberculosis", meta_fields=["geo_accession","series_id","characteristics_ch1", "extract_protocol_ch1", "source_name_ch1", "title"])
lung_TB = a4.meta.meta(file, "lung tuberculosis", meta_fields=["geo_accession","series_id","characteristics_ch1", "extract_protocol_ch1", "source_name_ch1", "title"])

lung_TB2 = a4.meta.meta(file, "lung TB", meta_fields=["geo_accession","series_id","characteristics_ch1", "extract_protocol_ch1", "source_name_ch1", "title"])
```

```python
import pandas as pd
TB_meta_updated= pd.concat([PTB_meta, activeTB_meta, MTB_meta, MTB_meta2, lung_TB, lung_TB2], ignore_index=True)
TB_meta_updated
```

```python
TB_meta_list_updated = TB_meta_updated["series_id"].unique().tolist()
print(TB_meta_updated['series_id'].unique())
print(len(TB_meta_updated['series_id'].unique()))
```

```python
TB_meta_split = []
for item in TB_meta_list_updated:
    TB_meta_split.extend([x.strip() for x in item.split(',')])
TB_meta_split = np.array(TB_meta_split)

print(TB_meta_split)
print(len(TB_meta_split))

# print the dup
print("Duplicates in TB_meta_split:")
duplicates = set(x for x in TB_meta_split if TB_meta_split.tolist().count(x) > 1)
print(duplicates)

# remove duplicates
TB_meta_split = np.unique(TB_meta_split)
print("After removing duplicates:")
print(len(TB_meta_split))
```
