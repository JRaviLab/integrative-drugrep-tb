```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
import data_funcs
import correlation_funcs
```

```python
drug_data, landmark_genes = data_funcs.read_drug_data()
dis_data = data_funcs.read_dis_data(landmark_genes)
```

```python
dis_sample = dis_data.iloc[:,0]

dis_sample.dropna(inplace=True)
dis_sample.index
```

```python
dis_sample = dis_data.iloc[:,0]

dis_sample.dropna(inplace=True)
drug_data_nona = drug_data.loc[dis_sample.index]
lasso = Lasso(alpha=0.1, max_iter=10000)
lasso.fit(drug_data_nona.values, dis_sample.values)
# output.loc[drug_data_nona.index, 'lasso'] = lasso.coef_
```

```python
# NUMBER 2

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Predict using the trained Lasso model
predicted_values = lasso.predict(drug_data_nona.values)

# Calculate R-squared score
r2 = round(r2_score(dis_sample.values, predicted_values), 4)

# Calculate Mean Absolute Error (MAE)
mae = round(mean_absolute_error(dis_sample.values, predicted_values), 4)

# Calculate Mean Squared Error (MSE)
mse = round(mean_squared_error(dis_sample.values, predicted_values), 4)

# Calculate Root Mean Squared Error (RMSE)
rmse = round(mean_squared_error(dis_sample.values, predicted_values, squared=False), 4)

print("R-squared score:", r2)
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
```
